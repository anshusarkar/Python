```python
import folium 

world_map = folium.Map()

world_map

india_map = folium.Map(location = [20.5937, 78.9629], zoom_start=4) # Additional parameter tiles="Stamen Toner" can be added that can help to look into the map with diffrent styles

india_map 

# Create a feture group

kolkata = folium.map.FeatureGroup()

kolkata.add_child(
    folium.features.CircleMarker([22.5726, 88.3639], radius = 5, color = 'red', fill_color = "Red")

)
india_map.add_child(kolkata)

# Adding a marker over kolkata

folium.Marker([22.5726, 88.3639], popup='Kolkata').add_to(india_map)

india_map
```
