```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('categorical_vehicle_sales_data.csv')
recession_years = [2008, 2009]
```

```python
# Define recession years (example)
recession_years = [2010, 2011]

plt.figure(figsize=(12, 8))  # Adjust figsize to reduce image size
plt.plot(data['Year'], data['Splinder'], marker='o', color='blue', linestyle='-', label='Splinder')
plt.plot(data['Year'], data['Apache'], marker='o', color='green', linestyle='-', label='Apache')
plt.plot(data['Year'], data['Kawasaki'], marker='o', color='yellow', linestyle='-', label='Kawasaki')
plt.plot(data['Year'], data['Mustang'], marker='o', color='orange', linestyle='-', label='Mustang')
plt.plot(data['Year'], data['Lamborgini'], marker='o', color='green', linestyle='-', label='Lamborgini')
plt.plot(data['Year'], data['Ford'], marker='o', color='pink', linestyle='-', label='Ford')

# Annotate recession years

for year in recession_years:
    plt.axvline(x=year, color='red', linestyle='--', linewidth=0.5)
    plt.text(year,10 , 'Recession', ha='center', color='oragne', fontsize=10)
    plt.text(year,100 , 'Recession', ha='center', color='red', fontsize=10)
      

plt.title('Vehicle Sales Throughout Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True) 
plt.legend()
#plt.show()
plt.savefig('Line_plot_2.png', dpi=300) 
```
