```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('automobile_sales.csv')
```

```python
# Plotting the line plot
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(df['Year'], df['Sales'], marker='o', color='blue', linestyle='-')  # Plotting the line
plt.title('Automobile Sales Fluctuation')  # Adding a title
plt.xlabel('Year')  # Adding label to the x-axis
plt.ylabel('Sales')  # Adding label to the y-axis
plt.grid(True)  # Adding gridlines
plt.xticks(df['Year'], rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.savefig('Line_plot_1.png', dpi=300) 
plt.show()  # Display the plot
```
