```python
import matplotlib.pyplot as plt

# Data
expenditure = [30, 40]  # Advertising expenditure during recession and non-recession periods, respectively
periods = ['Recession', 'Non-Recession']

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(expenditure, labels=periods, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen'], explode=(0, 0.1))
plt.title('Advertising Expenditure of XYZAutomotives')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.savefig('Pie_1.png')
plt.show()
```
