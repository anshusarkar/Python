```python
import matplotlib.pyplot as plt
```

```python
average_veichicle_price = [99999, 12999, 13999, 98999, 14599]

years = [1999, 2000, 2001, 2002, 2003]

sales_volume_recession = [1000, 750, 610, 880, 380]
```

```python
# scatter plot

plt.scatter(average_veichicle_price, years, s = sales_volume_recession, alpha=0.5, color='blue')
plt.title('Scatter plot of sales')
plt.xlabel('Average veichicle price')
plt.ylabel('Year')
plt.savefig('Scatter.png')
plt.show()
```
