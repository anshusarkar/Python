```python
import matplotlib.pyplot as plt
```

```python
session_names = ["Spring", "Summer", "Autumn", "Winter"]

automobile_sales_data = [1880, 1020, 1670, 2040]
```

```python
# Bubbuple plots i.e. the scatter plot 

# Plotting the Spring sales data

plt.scatter(session_names, automobile_sales_data, s = automobile_sales_data, alpha = 0.5, color = "green",)
plt.title('Impact on automobile sales in seasions')
plt.xlabel('seasion name')
plt.ylabel('automobile sales')
plt.savefig('Bubble.png')
plt.show()
```
