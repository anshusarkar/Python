```python
import matplotlib.pyplot as plt 
```

```python
quarters = ['Q1 2023', 'Q2 2023', 'Q3 2023', 'Q4 2023', 'Q1 2024', 'Q2 2024']
gdp_growth = [0.5, 0.7, 0.3, -0.2, -0.5, -0.3]  # Example GDP growth rates (%)
unemployment_rate = [5.2, 5.0, 5.5, 5.8, 6.2, 6.5]  # Example Unemployment rates (%)

# Plotting both GDP growth and Unemployment Rate on the same graph
plt.figure(figsize=(10, 5))

plt.plot(quarters, gdp_growth, marker='o', color='b', label='GDP Growth Rate')
plt.plot(quarters, unemployment_rate, marker='o', color='r', label='Unemployment Rate')

plt.title('Recession Report Statistics')
plt.xlabel('Quarters')
plt.ylabel('Rate (%)')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()  # Adding legend to differentiate between lines
plt.tight_layout()
plt.savefig('RecessionReportgraphs.png')
plt.show()
```
