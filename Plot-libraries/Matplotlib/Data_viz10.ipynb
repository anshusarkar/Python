```python
import matplotlib.pyplot as plt
```

```python
years = ['2019', '2020', '2021', '2022', '2023']  # Example years
gdp_growth = [2.3, -3.5, 4.2, 3.0, 2.5]  # Example GDP growth rates (%)
inflation_rate = [1.8, 2.5, 2.3, 2.8, 3.2]  # Example inflation rates (%)
unemployment_rate = [3.7, 6.2, 5.0, 4.5, 4.0]  # Example unemployment rates (%)

# Plotting GDP growth, Inflation Rate, and Unemployment Rate
plt.figure(figsize=(10, 5))

plt.plot(years, gdp_growth, marker='o', color='b', label='GDP Growth Rate')
plt.plot(years, inflation_rate, marker='o', color='r', label='Inflation Rate')
plt.plot(years, unemployment_rate, marker='o', color='g', label='Unemployment Rate')

plt.title('Yearly Report Statistics')
plt.xlabel('Year')
plt.ylabel('Rate (%)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('YearlyReportgraphs.png')
plt.show()
```
