```python
import matplotlib.pyplot as plt

# Sample GDP data (replace with your actual data)
years = range(2000, 2020) # That's made-up data of an imaginary country 

gdp_values = [5000, 5200, 5300, 5400, 5500, 5600, 5700, 5500, 5400, 5300, 
              5200, 5100, 5000, 4900, 4800, 4700, 4600, 4500, 4400, 4300]

# Sample recession periods (replace with your actual recession periods)
recession_periods = [(2001, 2002), (2008, 2009)]

# Create a figure and subplots
fig, axs = plt.subplots(len(recession_periods), 1, figsize=(10, 6), sharex=True)

# Plot GDP data for each period
for i, period in enumerate(recession_periods):
    start, end = period
    recession_years = range(start, end + 1)
    recession_gdp = [gdp_values[year - 2000] for year in recession_years] # Nice use of list comprehansion ! you know enreaching the list with data by an iteration of a loop ! 
    
    axs[i].plot(recession_years, recession_gdp, marker='o', linestyle='-', color='red')
    axs[i].set_title(f'Recession Period: {start}-{end}')
    axs[i].set_ylabel('GDP')
    axs[i].grid(True)

# Plot GDP data for non-recession period
non_recession_years = [year for year in years if year not in [year for period in recession_periods for year in range(period[0], period[1] + 1)]] # Too much use of list comprehansion ... 

# Insted of the code chuck above that has too much list comprehansion use the below line of code chunck ! to understand the code if u are felling lazy !

# non_recession_years = []
# for year in years:
#     is_recession_year = False
#     for period in recession_periods:
#         if year >= period[0] and year <= period[1]:
#             is_recession_year = True
#             break
#     if not is_recession_year:
#         non_recession_years.append(year)

non_recession_gdp = [gdp_values[year - 2000] for year in non_recession_years]

axs[-1].plot(non_recession_years, non_recession_gdp, marker='o', linestyle='-', color='blue')
axs[-1].set_title('Non-Recession Period')
axs[-1].set_ylabel('GDP')
axs[-1].grid(True)

# Set common X-axis label
plt.xlabel('Year')

# Adjust layout and display the plot
plt.tight_layout()
plt.savefig('Subplot.png')
plt.show()
```
