```python
import matplotlib.pyplot as  plt
```

```python
# Datasets

vehicles_sales_fig = [1245, 4325, 3956, 4567, 2345, 2312]

years = [ 2000, 2001, 2002, 2003, 2004, 2005]
```

```python
plt.figure(figsize=(8,8))
plt.plot(years,vehicles_sales_fig , marker='o', color='green', linestyle='-')
plt.text(2004, 2345, 'Re', color = 'red')
plt.text(2005, 2312, 'Re', color = 'red')
plt.xlabel("Years")
plt.ylabel("Vehicle sales")
plt.title("The effect of the unemployment rate on vehicle type and sales during the Recession Period")
plt.savefig('Line_plot_3.png')
plt.show()
```
