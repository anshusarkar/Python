```python
import matplotlib.pyplot as plt
```

```python
total_expenditure = [10, 12, 15, 25, 32, 44, 45, 45, 45, 51, 65, 70, 80, 90, 98]

total_expenditure.sort()

vehicles = [
    "Toyota Corolla",
    "Honda Civic",
    "Ford Focus",
    "Chevrolet Spark",
    "Nissan Versa",
    "Hyundai Accent",
    "Kia Rio",
    "Subaru Impreza",
    "Volkswagen Golf",
    "Mazda3",
    "Audi A3",
    "BMW 3 Series",
    "Mercedes-Benz C-Class",
    "Tesla Model 3",
    "Ferrari 488 Spider"
]

# List of vehicles sorted from least likely to require a commercial to most likely to require a commercial

colors = [
    "red",
    "blue",
    "green",
    "yellow",
    "orange",
    "purple",
    "pink",
    "cyan",
    "magenta",
    "teal",
    "lime",
    "brown",
    "gray",
    "black",
    "chocolate",
]
# List of colors o represent pie chart
```

```python
# pie plot
plt.figure(figsize = (8, 8))
plt.pie(total_expenditure,labels = vehicles, autopct='%1.1f%%', startangle=140, colors=colors, explode=[0] * len(total_expenditure))
plt.title('Pie_chart_showing_vehicles_that_required_least_expenditure_to_most_in_recession')
plt.savefig('Pie_2.png')
plt.show()
```
